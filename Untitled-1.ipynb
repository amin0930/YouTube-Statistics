```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
```

```python
df = pd.read_csv('Global YouTube Statistics.csv', encoding='latin-1')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(df.loc[(df.Longitude.isna()) & (df.Latitude.isna())].index.tolist(), inplace=True)
df.reset_index().drop(columns=['index'], inplace=True)
df.shape
```

```python
top_channels_by_subscribers = df.head(5)
top_channels_by_subscribers
```

```python
highest_yearly_earning_of_top_five_subscribers = top_channels_by_subscribers['highest_yearly_earnings']
fig1, ax1 = plt.subplots(figsize=(10,5))
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)

highest_yearly_earning_of_top_five_subscribers.plot(kind='bar', ax=ax1, color='green', alpha=1 )
ax1.set_xticks([0,1,2,3,4])
ax1.set_xticklabels(labels=['T-Series', 'YouTube Movies', 'MrBeast', 'Cocomelon-Nursery Rhymes', 'SET India'], rotation = 0, fontsize=9, color='black')
ax1.set_xlabel('Youtuber', fontsize = 15, color='gray')
ax1.set_ylabel('yearly earnings',fontsize= 15,color = 'gray')
ax1.set_facecolor('silver')
fig1.set_facecolor('silver')
plt.ylim(0,130000000)
custom_y_ticks = [0, 20000000, 40000000, 60000000, 80000000, 100000000,120000000]
custom_y_labels = ['0', '20M$', '40M$', '60M$', '80M$', '100M$', '120M$']
plt.yticks(custom_y_ticks, custom_y_labels)
ax1.annotate('170M Subs - Almost 0 earning!', xy=(1,2000000), xytext=(0.4,30000000), arrowprops=dict(arrowstyle="->", connectionstyle='arc3,rad=-0.3', color ='red'), color='red',fontsize=8)
ax1.annotate('245M Subs - Music Category', xy=(0.25,100000000), xytext=(0.1,115000000), arrowprops=dict(arrowstyle = '->', connectionstyle='arc3, rad=-0.3', color = 'blue'), color = 'blue', fontsize=8)
ax1.annotate('166M Subs - Entertainment Category', xy=(2.,65000000), xytext=(1,80000000),arrowprops=dict(arrowstyle = '->', connectionstyle='arc3, rad=-0.3', color = 'blue'), color = 'blue', fontsize=8)
ax1.annotate('162M Subs - Education Category', xy=(3, 97000000), xytext=(2.2,107000000), arrowprops=dict(arrowstyle = '->', connectionstyle='arc3, rad=-0.3', color = 'blue'), color = 'blue', fontsize=8)
ax1.annotate('159M Subs - Entertainment Category', xy=(4.1,90000000), xytext=(3.2,98000000),arrowprops=dict(arrowstyle = '->', connectionstyle='arc3, rad=-0.3', color = 'blue'), color = 'blue', fontsize=8 )
ax1.set_title('What is the highest annual earning of top 2023 channels since their creation?', color='teal',fontsize=15)
```

```python
fig3, ax3 = plt.subplots(figsize= (10,5))
correlation_highest_yearly_income_subscribers = df['highest_yearly_earnings'].corr(df['subscribers'])
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
sns.regplot(x = "highest_yearly_earnings",
            y = "subscribers", 
            ci = None,
            data = df,)
ax3.set_facecolor('silver')
fig3.set_facecolor('silver')
ax3.annotate('correlation  = 0.44', xy=(120000000, 150000000),fontsize=14,color = 'navy')
ax3.set_title('Any Correlation between the number of subscribers and the highest annual earning of channels since their creaion?', fontsize=10, color = 'teal')
```
